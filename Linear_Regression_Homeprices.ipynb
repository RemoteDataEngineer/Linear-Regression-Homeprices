```python
### Step 1. Import Libraries
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
### Step 2. Training Dataset
```

```python
df = pd.read_csv("homeprices.csv") # train
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('prince(US$)')
plt.scatter(df.area, df.price, color = 'red', marker = '+')
```

```python
### Step 3. Test Dataset
```

```python
d = pd.read_csv('areas.csv')
d
```

```python
### Step 4. Create linear regression model with Train dataset containing area and price
```

```python
#create linear regression
reg = linear_model.LinearRegression()

reg.fit(df[['area']], df.price)
```

```python
#slope of the line of the regression equation
reg.coef_
```

```python
#Y intercept
```

```python
reg.intercept_
```

```python
#use a row in the features column to get an output
```

```python
### y = m*x+b
### output or dependant variable  = slope * feature + Y intercept
### price  = m * area + b
y =  reg.coef_ * 3300 + reg.intercept_
y
```

```python
135.78767123 * 3300  + 180616.43835616432
```

```python
### Step 5. Use model on test dataset to get predictions for homeprices
```

```python
p = reg.predict(d)
```

```python
d['prices'] = p
d
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('prince(US$)')
plt.scatter(d.area, d.prices, color = 'red', marker = '+')
```
